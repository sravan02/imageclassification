```python
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
import numpy as np
import os
```

```python
link = 'https://www.amazon.in/New-Apple-iPhone-Pro-512GB/product-reviews/B08L5WVZJ7/ref=cm_cr_dp_d_show_all_btm?ie=UTF8&reviewerType=all_reviews'
```

```python
page = requests.get(link)
page
```

```python
page.content
```

```python
Soup = bs(page.content,'html.parser')
```

```python
print(Soup.prettify())
```

```python
names = Soup.find_all('span',class_='a-profile-name')
```

```python
names
```

```python
cust_name = []
for i in range(0,len(names)):
  cust_name.append(names[i].get_text())
cust_name
```

```python
cust_name.pop(0)
```

```python
cust_name.pop(0)
```

```python
title = Soup.find_all('a',class_='review-title-content')
```

```python
title
```

```python
review_title = []
for i in range(0,len(title)):
  review_title.append(title[i].get_text())
review_title
```

```python
review_title[:] = [titles.lstrip('\n') for titles in review_title]
review_title
```

```python
review_title[:] = [titles.rstrip('\n') for titles in review_title]
review_title
```

```python
rating = Soup.find_all('i',class_='review-rating')
rating
```

```python
rate = []
for i in range(0,len(rating)):
  rate.append(rating[i].get_text())
rate
```

```python
rate.pop(0)
```

```python
rate.pop(0)
```

```python
review = Soup.find_all('span',{"data-hook":"review-body"})
review
```

```python
review_contant = []
for i in range(0,len(review)):
  review_contant.append(review[i].get_text())
review_contant
```

```python
review_contant[:] = [reviews.lstrip('\n\n') for reviews in review_contant]
review_contant
```

```python
review_contant[:] = [reviews.rstrip('\n\n') for reviews in review_contant]
review_contant
```

```python
cust_name
review_title
rate
review_contant
```

```python
df = pd.DataFrame()
```

```python
df['Customer Name']=cust_name
df['Rating']=rate
df['Review title']=review_title
df['Reviews']=review_contant
```
